# Exactitud vs. Precisión
## Exactitud

Es que tan cercano se esta del valor verdadero, es decir si nuestro valor por ejemplo es: `2` entonces `1.8` es mas exacto que `1.5` por que `1.8` esta mas cerca de 2.

## Precisión

Es que dan desviado esta cada valor de los demas valores sin contar con el valor verdadero, es decir por ejemplo:
`1.5, 1.7, 1.6, 1.4, 1.8, 1.7` es mas preciso que `1.5, 5.6, 1.0, 79.5, 56.6`


# Tipos de Errores
Un error es la diferencia entre el valor verdadero con el valor y la aproximación que obtenemos al calcular o medir algo.
