# Exactitud vs. Precisión
## Exactitud
Es que tan cercano se esta del valor verdadero, es decir si nuestro valor por ejemplo es: `2` entonces `1.8` es mas exacto que `1.5` por que `1.8` esta mas cerca de 2.

## Precisión
Es que dan desviado esta cada valor de los demas valores sin contar con el valor verdadero, es decir por ejemplo:
`1.5, 1.7, 1.6, 1.4, 1.8, 1.7` es mas preciso que `1.5, 5.6, 1.0, 79.5, 56.6`

# Tipos de Errores
Un error es la diferencia entre el valor verdadero con el valor y la aproximación que obtenemos al calcular o medir algo.

## Error verdadero
Es la diferencia entre el valor medido(aproximacion) y el valor verdadero $$E_v = V_v - V_a$$ donde $E_v$ es el Error verdadero, $V_v$ es el Valor verdadero y $V_a$ es el Valor aproximado o medido, este tambien nos permite saber si el error fue al lado izquierdo o al lado positivo con ayuda del signo.   
Esto seria la funcion en python para calcularlo:


```python
def error_verdadero(valor_verdadero, valor_aproximado):
  return valor_verdadero - valor_aproximado
```

## Error Absoluto
Cuando calculamos esto no nos importa la direccion a la que el error se nos fue, solo nos importa la magnitud de esta, es solo el error verdadero ignorando el signo o calculando su valor absoluto

```python
def error_absoluto(valor_verdadero, valor_aproximado):
    return abs(error_verdadero(valor_verdadero, valor_aproximado))
```

## Error Relativo y porcentual
Este error se usa para calcular que tan significante es el error que se cometio, ejemplo:
si se fallo por 1 metro en una distancia de 3 metros esto es muy significante es decir el error es alto, de mas o menos el 33.3%.
si se fallo por 1 metro en una distancia de 10000 metros ese 1 metro no es muy significante, es decir el error es bajo, de mas o menos de 0.999 %
### Error relativo
$$E_r = \frac{|V_v - V_a|}{|V_v|}$$

El error absoluto dividido entre el absoluto del valor verdadero

### Error relativo porcentual
$$E_{rp} = E_r \times 100\%$$
El error relativ x 100 , y se expresa con el simbolo de `%` al final.



```python
def error_relativo(valor_verdadero, valor_aproximado):
  if (valor_verdadero == 0):
    raise ValueError("El valor verdadero no puede ser cero")
  else :
    return error_absoluto(valor_verdadero, valor_aproximado)/abs(valor_verdadero)
```

```python
def error_relativo_porcentual(valor_verdadero, valor_aproximado):
  return error_relativo(valor_verdadero, valor_aproximado)*100
```

# Metodos cerrados
cuando no se puede encontrar la solucion a una ecuacion, pero si se puede conocer el rango en el que se encuentra la solucion, podemos aplicar un metodo cerrado para poder allar su solucion

## Método de la Bisección

```python
def metodo_de_la_biseccion(ec, a, b, tolerancia):
    fa = ec(a)
    fb = ec(b)

    if fa * fb > 0:
        print("a y b no contienen el resultado")
        return None
    m = (a + b) / 2
    fm = ec(m)

    if fm == 0:
        return m

    if abs(b - a) / 2 < tolerancia:
        return m

    if fa * fm < 0:
        return metodo_de_la_biseccion(ec, a, m, tolerancia)
    else:
        return metodo_de_la_biseccion(ec, m, b, tolerancia)
```

## Método de la Falsa Posición

```python
def falsa_posicion (ec, a, b, tolerancia, xo=None):
  fa = ec(a)
  fb = ec(b)

  if fa * fb > 0:
      print("a y b no contienen el resultado")
      return None
  x = b-(fb*(a-b))/(fa-fb)
  fx = ec(x)
  if fx == 0:
    return x

  if xo != None and abs(x - xo) < tolerancia:
      return x

  if fa * fx < 0:
      return falsa_posicion(ec, a, x, tolerancia, x)
  else:
      return falsa_posicion(ec, x, b, tolerancia, x)
```

# Métodos Abiertos

```python
def derivar(ecuacion, orden=1):
    x = sp.Symbol('x')
    g_simbolica = sp.sympify(ecuacion)
    derivada = sp.diff(g_simbolica, x, orden)
    return sp.lambdify(x, derivada, 'math')
```

## Método de Punto Fijo

```python
def punto_fijo(g, g_texto, x_actual, tolerancia, max_iteraciones, iteracion=1):
  if iteracion == 1:
    dg = derivar(g_texto)
    e = dg(x_actual)
    if abs(e) >= 1:
      print(f"La derivada es {e}, lo que nos indica que va a divergir")
      return None

  if iteracion > max_iteraciones:
      print("maximo de iteraciones alcanzado")
      return None

  x_nuevo = g(x_actual)

  if abs(x_nuevo - x_actual) < tolerancia:
      return x_nuevo

  return punto_fijo(g, g_texto, x_nuevo, tolerancia, max_iteraciones, iteracion + 1)
```

## Método de Newton-Raphson
Fundamental. Utiliza la primera derivada de la función para trazar tangentes. Convergencia cuadrática.
Requiere acceso analítico a la derivada. A nivel de sistema, evalúa dos funciones en la FPU por ciclo.

```python
def newton_raphson(ec, ec_texto, x_actual, tolerancia, max_iteraciones, iteracion=1):
    if iteracion > max_iteraciones:
        print("maximo de iteraciones alcanzado")
        return None


    df = derivar(ec_texto, 1)

    fx = ec(x_actual)
    dfx = df(x_actual)

    if dfx == 0:
        print("División por, no se puede")
        return None

    x_nuevo = x_actual - (fx / dfx)

    if abs(x_nuevo - x_actual) < tolerancia:
        return x_nuevo

    return newton_raphson(ec, ec_texto, x_nuevo, tolerancia, max_iteraciones, iteracion + 1)

```

## Método de la Secante
Aproximación de Newton mediante diferencias finitas. No exige derivadas analíticas.
Ideal cuando la función es una caja negra u obtener el diferencial colapsa la memoria.
Requiere dos valores iniciales.

```python
def secante(ec, x_anterior, x_actual, tolerancia, max_iteraciones, iteracion=1):
    if iteracion > max_iteraciones:
        print("maximo de iteraciones alcanzado")
        return None

    fx_actual = ec(x_actual)
    fx_anterior = ec(x_anterior)


    denominador = fx_anterior - fx_actual
    if denominador == 0:
         print("División por cero")
         return None

    x_nuevo = x_actual - (fx_actual * (x_anterior - x_actual)) / denominador

    if abs(x_nuevo - x_actual) < tolerancia:
        return x_nuevo

    return secante(ec, x_actual, x_nuevo, tolerancia, max_iteraciones, iteracion + 1)

```

# Métodos de Orden Superior y Polinomios

## Método de Halley
Newton-Raphson inyectado con esteroides. Usa la segunda derivada para lograr una convergencia cúbica (p=3).
Alto costo computacional: Tres llamadas a la FPU por iteración. Útil para raíces múltiples.
Fórmula: x_{n+1} = x_n - (2 * f(x) * f'(x)) / (2 * [f'(x)]^2 - f(x) * f''(x))

```python
def halley(ec, ec_texto, x_actual, tolerancia, max_iteraciones, iteracion=1):
    if iteracion > max_iteraciones:
        print("maximo de iteraciones alcanzado.")
        return None

    df = derivar(ec_texto, 1)
    ddf = derivar(ec_texto, 2)

    fx = ec(x_actual)
    dfx = df(x_actual)
    ddfx = ddf(x_actual)

    denominador = 2 * (dfx**2) - (fx * ddfx)

    if denominador == 0:
        print("division con 0, no se puede")
        return None

    x_nuevo = x_actual - ((2 * fx * dfx) / denominador)

    if abs(x_nuevo - x_actual) < tolerancia:
        return x_nuevo

    return halley(ec, ec_texto, x_nuevo, tolerancia, max_iteraciones, iteracion + 1)
```

## Método de Muller
Utiliza una parábola (polinomio de segundo grado) interpolada en tres puntos.
Permite descubrir raíces complejas operando fuera del rango de los números reales estándar,
gestionando desbordamientos nativos usando la librería cmath (Complex Math).

```python
def muller(ec, x0, x1, x2, tolerancia, max_iteraciones, iteracion=1):
    if iteracion > max_iteraciones:
        print("maximo de iteraciones alcanzado")
        return None

    h0 = x1 - x0
    h1 = x2 - x1

    if h0 == 0 or h1 == 0 or (x2 - x0) == 0:
         print("no se puede")
         return None

    d0 = (ec(x1) - ec(x0)) / h0
    d1 = (ec(x2) - ec(x1)) / h1

    a = (d1 - d0) / (x2 - x0)
    b = a * h1 + d1
    c = ec(x2)

    rad = cmath.sqrt(b**2 - 4*a*c)


    denominador_suma = b + rad
    denominador_resta = b - rad

    if abs(denominador_suma) > abs(denominador_resta):
        denominador_optimo = denominador_suma
    else:
        denominador_optimo = denominador_resta

    if denominador_optimo == 0:
        print("Muller: Denominador cero.")
        return None

    x3 = x2 - (2 * c) / denominador_optimo

    if abs(x3 - x2) < tolerancia:

        if abs(x3.imag) < 1e-12:
             return x3.real
        return x3

    return muller(ec, x1, x2, x3, tolerancia, max_iteraciones, iteracion + 1)
```
