# Exactitud vs. Precisión
## Exactitud

Es que tan cercano se esta del valor verdadero, es decir si nuestro valor por ejemplo es: `2` entonces `1.8` es mas exacto que `1.5` por que `1.8` esta mas cerca de 2.

## Precisión

Es que dan desviado esta cada valor de los demas valores sin contar con el valor verdadero, es decir por ejemplo:
`1.5, 1.7, 1.6, 1.4, 1.8, 1.7` es mas preciso que `1.5, 5.6, 1.0, 79.5, 56.6`


# Tipos de Errores
Un error es la diferencia entre el valor verdadero con el valor y la aproximación que obtenemos al calcular o medir algo.

## Error verdadero
Es la diferencia entre el valor medido(aproximacion) y el valor verdadero $$E_v = V_v - V_a$$ donde $E_v$  es el Error verdadero, $V_v$ es el Valor verdadero y $V_a$ es el Valor aproximado o medido, este tambien nos permite saber si el error fue al lado izquierdo o al lado positivo con ayuda del signo.   
Esto seria la funcion en python para calcularlo:

```python
def error_verdadero(valor_verdadero, valor_aproximado):
  return valor_verdadero - valor_aproximado
```

## Error Absoluto
Cuando calculamos esto no nos importa la direccion a la que el error se nos fue, solo nos importa la magnitud de esta, es solo el error verdadero ignorando el signo o calculando su valor absoluto.

```python
def error_absoluto(valor_verdadero, valor_aproximado):
	return abs(error_verdadero(valor_verdadero, valor_aproximado))
```

## Error Relativo y porcentual
Este error se usa para calcular que tan significante es el error que se cometio, ejemplo:
si se fallo por 1 metro en una distancia de 3 metros esto es muy significante es decir el error es alto, de mas o menos el 33.3%.
si se fallo por 1 metro en una distancia de 10000 metros ese 1 metro no es muy significante, es decir el error es bajo, de mas o menos de 0.999 %
### Error relativo
$$E_r = \frac{|V_v - V_a|}{|V_v|}$$

El error absoluto dividido entre el absoluto del valor verdadero

### Error relativo porcentual
$$E_{rp} = E_r \times 100\%$$
El error relativ x 100 , y se expresa con el simbolo de `%` al final.


```python
# error relativo
def error_relativo(valor_verdadero, valor_aproximado):
  if (valor_verdadero == 0):
    raise ValueError("El valor verdadero no puede ser cero")
  else :
    return error_absoluto(valor_verdadero, valor_aproximado)/abs(valor_verdadero)

# error relativo porcentual
def error_relativo_porcentual(valor_verdadero, valor_aproximado):
  return error_relativo(valor_verdadero, valor_aproximado)*100
```

# Metodos cerrados
cuando no se puede encontrar la solucion a una ecuacion, pero si se puede conocer el rango en el que se encuentra la solucion, podemos aplicar un metodo cerrado para poder allar su solucion

## Método de la Bisección
Consiste en agarrar dos numeros (los extremos del rango) que deben cumplir con la condicion de que al evaluar su resultado sea uno positivo y el otro negativo, calcular el punto medio, evaluar la funcion con la variable igual a los extremos y al punto medio, es decir si tus extremos son `a` y `b`
$$f(a) = negativo $$

$$f\left(\frac{a+b}{2}\right) = \text{positivo}$$
$$f(b) = positivo$$ entonces tu nuevo rango seria entre a y $\frac{a+b}{2}$, y vuelves a repetir este metodo, una y otra vez hasta que se cumpla con la tolerancia.


```python

def metodo_de_la_biseccion(ec, a, b, tolerancia):
    fa = ec(a)
    fb = ec(b)

    if fa * fb > 0:
        print("a y b no contienen el resultado")
        return None
    m = (a + b) / 2
    fm = ec(m)

    if fm == 0:
        return m

    if abs(b - a) / 2 < tolerancia:
        return m

    if fa * fm < 0:
        return metodo_de_la_biseccion(ec, a, m, tolerancia)
    else:
        return metodo_de_la_biseccion(ec, m, b, tolerancia)
```

## Método de la Falsa Posición
Al igual que el anterior este requiere dos numeros que contengan el resultado, pero que a diferencia de biseccion que trata usando el punto medio aqui, apunta usando una linea secante
$$x_r = b - \frac{f(b)(a - b)}{f(a) - f(b)}$$



```python
def falsa_posicion (ec, a, b, tolerancia, xo=None):
  fa = ec(a)
  fb = ec(b)

  if fa * fb > 0:
      print("a y b no contienen el resultado")
      return None
  x = b-(fb*(a-b))/(fa-fb)
  fx = ec(x)
  if fx == 0:
    return x

  if xo != None and abs(x - xo) < tolerancia:
      return x

  if fa * fx < 0:
      return falsa_posicion(ec, a, x, tolerancia, x)
  else:
      return falsa_posicion(ec, x, b, tolerancia, x)
```
